```python
#Task 1:Simple Interest using Function

def simple_interest(p, r, t):
    si = (p * r * t) / 100
    return si

print("Simple Interest:", simple_interest(1000, 5, 2))
```

```python
#Task 2:Reverse Array using Recursion

def reverse_array(arr, start, end):
    if start >= end:
        return
    arr[start], arr[end] = arr[end], arr[start]
    reverse_array(arr, start + 1, end - 1)

numbers = [1, 2, 3, 4, 5]
reverse_array(numbers, 0, len(numbers) - 1)
print("Reversed Array:", numbers)
```

```python
#Task 3:OOPS Example

class Student:
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks

    def display(self):
        print("Name:", self.name)
        print("Marks:", self.marks)

s1 = Student("Nandini", 85)
s1.display()
```

```python
#Task 4:Exception Handling Example
try:
    a = int(input("Enter a number: "))
    b = int(input("Enter another number: "))
    print("Division:", a / b)
except ZeroDivisionError:
    print("Cannot divide by zero!")
except ValueError:
    print("Invalid input!")
finally:
    print("Program executed.")
```
