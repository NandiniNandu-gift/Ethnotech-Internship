```python
#Task 1 – Students Marks > 80

import sqlite3

conn = sqlite3.connect("college.db")
cursor = conn.cursor()

cursor.execute("""
CREATE TABLE IF NOT EXISTS students (
    id INTEGER PRIMARY KEY,
    name TEXT,
    department TEXT,
    marks INTEGER
)
""")

cursor.execute("DELETE FROM students")  # Clear old data
cursor.executemany("""
INSERT INTO students (id, name, department, marks)
VALUES (?, ?, ?, ?)
""", [
    (1, 'Anu', 'CSE', 85),
    (2, 'Ravi', 'ECE', 72),
    (3, 'Kiran', 'CSE', 90),
    (4, 'Meena', 'EEE', 65),
    (5, 'Arjun', 'CSE', 88)
])

cursor.execute("SELECT name, marks FROM students WHERE marks > 80")

print("Students with marks > 80:")
for row in cursor.fetchall():
    print(row)

conn.commit()
conn.close()
```

```python
#Task 2 – Count Students by Department

import sqlite3

conn = sqlite3.connect("college.db")
cursor = conn.cursor()

cursor.execute("""
SELECT department, COUNT(*) 
FROM students
GROUP BY department
""")

print("Students count by department:")
for row in cursor.fetchall():
    print(row)

conn.close()
```

```python
#Task 3:Top 5 Highest Salary Employees
import sqlite3

conn = sqlite3.connect("mydatabase.db")
cursor = conn.cursor()

cursor.execute("""
CREATE TABLE IF NOT EXISTS employees (
    id INTEGER PRIMARY KEY,
    name TEXT,
    department TEXT,
    salary INTEGER
)
""")

cursor.execute("DELETE FROM employees")

cursor.executemany("""
INSERT INTO employees (id, name, department, salary)
VALUES (?, ?, ?, ?)
""", [
    (1, 'Rahul', 'HR', 25000),
    (2, 'Priya', 'IT', 50000),
    (3, 'Vikram', 'IT', 60000),
    (4, 'Divya', 'Finance', 45000),
    (5, 'Suresh', 'HR', 30000),
    (6, 'Lakshmi', 'IT', 70000),
    (7, 'Arun', 'Finance', 55000)
])

# Query
cursor.execute("""
SELECT name, salary 
FROM employees 
ORDER BY salary DESC 
LIMIT 5
""")

print("Top 5 Highest Salary Employees:")
for row in cursor.fetchall():
    print(row)

conn.commit()
conn.close()
```
