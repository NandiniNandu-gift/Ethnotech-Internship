```python
#Task 1: Create Students Table with Constraints
import sqlite3

conn = sqlite3.connect("CollegeDB.db")
cursor = conn.cursor()

cursor.execute("""
CREATE TABLE IF NOT EXISTS Students (
    student_id INTEGER PRIMARY KEY,
    name TEXT NOT NULL,
    email TEXT UNIQUE,
    age INTEGER CHECK(age >= 18),
    city TEXT DEFAULT 'Bangalore'
)
""")

cursor.execute("INSERT INTO Students (student_id, name, email, age) VALUES (1, 'Nandini', 'nandini@gmail.com', 20)")
conn.commit()

for row in cursor.execute("SELECT * FROM Students"):
    print(row)
```

```python
#Task 2: Create Courses and Enrollments (Foreign Key)

cursor.execute("PRAGMA foreign_keys = ON")

cursor.execute("""
CREATE TABLE IF NOT EXISTS Courses (
    course_id INTEGER PRIMARY KEY,
    course_name TEXT NOT NULL
)
""")

cursor.execute("INSERT INTO Courses VALUES (101, 'Python')")

cursor.execute("""
CREATE TABLE IF NOT EXISTS Enrollments (
    enroll_id INTEGER PRIMARY KEY,
    student_id INTEGER,
    course_id INTEGER,
    FOREIGN KEY(student_id) REFERENCES Students(student_id),
    FOREIGN KEY(course_id) REFERENCES Courses(course_id)
)
""")

cursor.execute("INSERT INTO Enrollments VALUES (1, 1, 101)")
conn.commit()

for row in cursor.execute("SELECT * FROM Enrollments"):
    print(row)
```

```python
#Task 3: Composite Primary Key Example
cursor.execute("""
CREATE TABLE IF NOT EXISTS Library (
    book_id INTEGER,
    member_id INTEGER,
    issue_date TEXT,
    PRIMARY KEY (book_id, member_id)
)
""")

cursor.execute("INSERT INTO Library VALUES (1, 101, '2026-02-25')")
conn.commit()

for row in cursor.execute("SELECT * FROM Library"):
    print(row)
```

```python
#Task 4: Employees Table with UNIQUE and CHECK
cursor.execute("""
CREATE TABLE IF NOT EXISTS Employees (
    emp_id INTEGER PRIMARY KEY,
    emp_name TEXT NOT NULL,
    salary REAL CHECK(salary > 10000),
    phone TEXT UNIQUE
)
""")

cursor.execute("INSERT INTO Employees VALUES (1, 'Ravi', 25000, '9876543210')")
conn.commit()

for row in cursor.execute("SELECT * FROM Employees"):
    print(row)
```
