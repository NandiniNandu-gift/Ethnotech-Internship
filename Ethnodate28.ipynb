```python
#Task 1 – CREATE TABLE (DDL)

import sqlite3

conn = sqlite3.connect('college.db')
cursor = conn.cursor()

cursor.execute('''
CREATE TABLE IF NOT EXISTS Student (
    id INTEGER PRIMARY KEY,
    name TEXT,
    age INTEGER,
    course TEXT,
    email TEXT
)
''')
print("Task 1: Table 'Student' created successfully!")

conn.close()
```

```python
#Task 2 – INSERT RECORDS (DML)

import sqlite3

conn = sqlite3.connect('college.db')
cursor = conn.cursor()

cursor.execute("INSERT INTO Student VALUES (1, 'Nandini', 21, 'CSE', 'nandini@email.com')")
cursor.execute("INSERT INTO Student VALUES (2, 'Rahul', 22, 'ECE', 'rahul@email.com')")
cursor.execute("INSERT INTO Student VALUES (3, 'Priya', 20, 'ME', 'priya@email.com')")

conn.commit()
print("Task 2: 3 Records inserted successfully!")

conn.close()
```

```python
#Task 3 – SELECT RECORDS (DML)

import sqlite3

conn = sqlite3.connect('college.db')
cursor = conn.cursor()

cursor.execute("SELECT * FROM Student")
students = cursor.fetchall()

print("Task 3: All Students Data:")
for student in students:
    print(student)

conn.close()
```
