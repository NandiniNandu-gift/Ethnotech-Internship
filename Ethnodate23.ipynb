```python
#Task1:Create Database and Table
import sqlite3

conn = sqlite3.connect("CompanyDB.db")
cursor = conn.cursor()

cursor.execute("""
CREATE TABLE IF NOT EXISTS Employee (
    emp_id INTEGER PRIMARY KEY,
    emp_name TEXT NOT NULL,
    salary INTEGER NOT NULL
)
""")

conn.commit()
conn.close()

print("Task 1 Completed: Database and Table Created")
```

```python
#Task 2: Insert Multiple Records

import sqlite3

conn = sqlite3.connect("CompanyDB.db")
cursor = conn.cursor()

cursor.execute("INSERT INTO Employee VALUES (1, 'Nandini', 30000)")
cursor.execute("INSERT INTO Employee VALUES (2, 'Rahul', 35000)")
cursor.execute("INSERT INTO Employee VALUES (3, 'Sneha', 40000)")

conn.commit()
conn.close()

print("Task 2 Completed: Records Inserted")
```

```python
#Task 3: Display All Records

import sqlite3

conn = sqlite3.connect("CompanyDB.db")
cursor = conn.cursor()

cursor.execute("SELECT * FROM Employee")
rows = cursor.fetchall()

for row in rows:
    print(row)

conn.close()
```

```python
#Task 4: Update Salary of an Employee

import sqlite3

conn = sqlite3.connect("CompanyDB.db")
cursor = conn.cursor()

cursor.execute("UPDATE Employee SET salary = 45000 WHERE emp_id = 2")

conn.commit()
conn.close()

print("Task 4 Completed: Salary Updated")
```

```python
#Task 5: Delete a Record

import sqlite3

conn = sqlite3.connect("CompanyDB.db")
cursor = conn.cursor()

cursor.execute("DELETE FROM Employee WHERE emp_id = 3")

conn.commit()
conn.close()

print("Task 5 Completed: Record Deleted")
```
