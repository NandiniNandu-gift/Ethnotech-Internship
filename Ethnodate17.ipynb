```python
#Task 1:Function Without Arguments

def greet():
    print("Hello, Welcome to Python Internship")

greet()
```

```python
#Task 2:Function With Arguments

def greet(name):
    print("Hello", name)

greet("Nandini")
```

```python
#Task 3:Function With Arguments and Return Type

def add(a, b):
    return a + b

result = add(10, 20)
print("Sum is:", result)
```

```python
#Task 4:Function With Default Argument

def greet(name="Student"):
    print("Hello", name)

greet()          
greet("Nandini")  
```
