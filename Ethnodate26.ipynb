```python
#TASK 1: Count Total Employees (Working in Anaconda)
import sqlite3

conn = sqlite3.connect("company.db")
cursor = conn.cursor()

cursor.execute("""
CREATE TABLE IF NOT EXISTS employees (
    emp_id INTEGER PRIMARY KEY,
    name TEXT,
    age INTEGER,
    department TEXT,
    salary REAL,
    joining_date TEXT
)
""")

cursor.execute("DELETE FROM employees")

employees_data = [
    (1, 'Nandini', 22, 'HR', 30000, '2023-01-15'),
    (2, 'Rahul', 25, 'IT', 45000, '2022-03-10'),
    (3, 'Sneha', 28, 'Finance', 50000, '2021-07-21'),
    (4, 'Arjun', 24, 'IT', 40000, '2023-06-01'),
    (5, 'Meera', 30, 'HR', 55000, '2020-11-11')
]

cursor.executemany("INSERT INTO employees VALUES (?, ?, ?, ?, ?, ?)", employees_data)

cursor.execute("SELECT COUNT(*) FROM employees")
print("Total Employees:", cursor.fetchone()[0])

conn.commit()
conn.close()
```

```python
#TASK 2: Average Salary
import sqlite3

conn = sqlite3.connect("company.db")
cursor = conn.cursor()

cursor.execute("SELECT AVG(salary) FROM employees")
print("Average Salary:", cursor.fetchone()[0])

conn.close()
```

```python
#TASK 3: Convert Names to Uppercase
import sqlite3

conn = sqlite3.connect("company.db")
cursor = conn.cursor()

cursor.execute("SELECT UPPER(name) FROM employees")

for row in cursor.fetchall():
    print(row[0])
```

```python
#TASK 4: Department-wise Total Salary
import sqlite3

conn = sqlite3.connect("company.db")
cursor = conn.cursor()

cursor.execute("""
SELECT department, SUM(salary) 
FROM employees
GROUP BY department
""")

for row in cursor.fetchall():
    print(row)

conn.close()#
```
