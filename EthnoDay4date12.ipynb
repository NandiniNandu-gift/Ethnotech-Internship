```python
#Task 1:Currency Converter

def currency_converter():
    amount = float(input("Enter the amount to convert: "))
    from_currency = input("Enter the currency to convert from (USD, EUR, INR): ").upper()
    to_currency = input("Enter the currency to convert to (USD, EUR, INR): ").upper()

    # Static conversion rates (Base: USD)
    rates = {
        "USD": 1,
        "EUR": 0.85,
        "INR": 75
    }

    if from_currency in rates and to_currency in rates:
        # Convert to USD first, then to target currency
        converted_amount = amount * (rates[to_currency] / rates[from_currency])
        print(f"{amount} {from_currency} = {converted_amount:.2f} {to_currency}")
    else:
        print("Invalid currency entered.")

currency_converter()
```
