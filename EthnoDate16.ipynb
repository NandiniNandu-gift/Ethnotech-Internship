```python
#Task 1: Conditional Statement 

a = 10
b = 25

if a > b:
    print("A is largest")
else:
    print("B is largest")
```

```python
#Task 2:Task 2: Loop Statement 

for i in range(1, 21):
    if i % 2 == 0:
        print(i)
```

```python
#Task 3: Jumping Statements 

for i in range(1, 10):
    if i == 5:
        break
    print(i)
```

```python
#Task 4: Match-Case Statement 

choice = 2

match choice:
    case 1:
        print("Addition")
    case 2:
        print("Subtraction")
    case 3:
        print("Multiplication")
    case _:
        print("Invalid Choice")
```
