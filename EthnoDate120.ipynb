```python
#Task 1:Simple Decorator 

def my_decorator(func):
    def wrapper():
        print("Before function execution")
        func()
        print("After function execution")
    return wrapper

@my_decorator
def greet():
    print("Hello Nandini")

greet()
```

```python
#Task 2:Iterator example

numbers = [10, 20, 30, 40]

# Create iterator
it = iter(numbers)

print(next(it))
print(next(it))
print(next(it))
print(next(it))
```

```python
#Task 3:Generator example

def numbers():
    yield 1
    yield 2
    yield 3

for num in numbers():
    print(num)
```
